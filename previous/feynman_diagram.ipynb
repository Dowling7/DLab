```python
from feynman import Diagram

# Create a diagram object
d = Diagram()

# Define the vertices
v1 = d.vertex([0.1, 0.25])  # Starting point for pi+
v2 = d.vertex([0.1, 0.75])  # Starting point for down quark
v3 = d.vertex([0.3, 0.5])   # Vertex where pi+ and d quark interact to form W+
v4 = d.vertex([0.9, 0.75])  # Ending point for muon neutrino
v5 = d.vertex([0.9, 0.25])  # Ending point for positively charged muon
v6 = d.vertex([0.7, 0.5])   # Vertex where W+ decays into muon neutrino and positively charged muon

# Add lines to the diagram

# pi+ to W+ interaction vertex
pi_line = d.line(v1, v3, style='dashed', label=r'$\pi^+$')

# Down quark to W+ interaction vertex
d_line = d.line(v2, v3, arrow=True, label='d')

# Explicit W+ boson line
W_boson = d.line(v3, v6, style='wiggly', label=r'$W^+$')

# W+ decaying to muon and neutrino
W_line_mu = d.line(v6, v5, arrow=True, label=r'$\mu^+$')
W_line_nu = d.line(v6, v4, arrow=True, label=r'$\nu_\mu$')

d.text(0.5,0.6, "W+-")
d.text(0.1,0.18, "d")
d.text(0.1,0.76, "Pi+-")

# Draw the diagram
d.draw()
```

```python
import matplotlib.pyplot as plt
from feynman import Diagram

# Setting up the Matplotlib figure
fig, ax = plt.subplots(figsize=(6, 6))
ax.axis('off')  # Turn off the axis
plt.title("Feynman Diagram Example")

# Initialize the Feynman diagram within the matplotlib axis
d = Diagram(ax=ax)

# Define the vertices for the Feynman diagram
v1 = d.vertex([0.1, 0.25])
v2 = d.vertex([0.1, 0.75])
v3 = d.vertex([0.5, 0.5])
v4 = d.vertex([0.9, 0.75])
v5 = d.vertex([0.9, 0.25])
v6 = d.vertex([0.7, 0.5])

# Draw the lines for the diagram
d.line(v1, v3, style='dashed', label=r'$\pi^+$')
d.line(v2, v3, arrow=True, label='d')
d.line(v3, v6, style='wiggly', label=r'$W^+$')
d.line(v6, v5, arrow=True, label=r'$\mu^+$')
d.line(v6, v4, arrow=True, label=r'$\nu_\mu$')

# Add text labels using Matplotlib's text function
ax.text(0.1, -0.1, "Initial state", ha='center')
ax.text(0.9, -0.1, "Final state", ha='center')
ax.text(0.5, 0.95, "Interaction", ha='center')

# Display the figure
plt.show()
```
